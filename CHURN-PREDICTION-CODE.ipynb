```python
#importing preliminary libraries

import pandas as pd
import numpy as np
```

```python
#                                     *****STEP 1: DATA PRERPOCESSING*****
```

```python
#loading the dataset

dataset = pd.read_csv("dataset.csv")
```

```python
#Encoding categorical variables (one hot encoding method)

df = pd.DataFrame(dataset)
dataset_enc1 = pd.get_dummies(data=df, columns=['Gender', 'Location'], drop_first=True)
```

```python
#Statistics of the numeric data

dataset_new = dataset.describe()
dataset_new
```

```python
#Encoding categorical variables using 

from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
dataset['Gender'] = label_encoder.fit_transform(dataset['Gender'])
dataset_enc2 = pd.get_dummies(dataset, columns=['Location'], drop_first=True)
```

```python
dataset_enc2
```

```python
# Split the data into features (X) and target (y)

X = dataset_enc2.drop(['Name', 'CustomerID', 'Churn'], axis=1)
y = dataset_enc2['Churn']
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
# Spliting data into training and testing sets

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state= 42)
X_train
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
# Standardizing numerical features (Age, Subscription_length_Months, Monthly_Bill, Total_Usage_GB)

scaler = StandardScaler()
X_train[['Age', 'Subscription_Length_Months', 'Monthly_Bill', 'Total_Usage_GB']] = scaler.fit_transform(X_train[['Age', 'Subscription_Length_Months', 'Monthly_Bill', 'Total_Usage_GB']])
X_test[['Age', 'Subscription_Length_Months', 'Monthly_Bill', 'Total_Usage_GB']] = scaler.transform(X_test[['Age', 'Subscription_Length_Months', 'Monthly_Bill', 'Total_Usage_GB']])
```

```python
X_train
```

```python
X_test
```

```python
#                                         *****STEP 2: FEATURE ENGNEERING*****
```

```python
# Calculating customer tenure in months

dataset_enc2['Customer_tenure_months'] = dataset_enc2['Subscription_Length_Months']
dataset_enc2['Customer_tenure_months']
```

```python
# Calculating average monthly usage

dataset_enc2['average_monthly_usage'] = dataset_enc2['Total_Usage_GB'] / dataset_enc2['Subscription_Length_Months']
dataset_enc2['average_monthly_usage'] 
```

```python
# Creating bins for age groups

bins = [0, 18, 30, 40, 50, 60, np.inf]
labels = ['<18', '18-30', '31-40', '41-50', '51-60', '60+']
dataset_enc2['Age_Group'] = pd.cut(dataset_enc2['Age'], bins=bins, labels=labels)
```

```python
dataset_enc2
```

```python
# Encoding age groups

dataset_enc3 = pd.get_dummies(dataset_enc2, columns=['Age_Group'], drop_first=True)
```

```python
dataset_enc3
```

```python
# Droping unnecessary columns

dataset_enc3.drop(['Name', 'CustomerID'], axis=1, inplace=True)
```

```python
dataset_enc3
```

```python
# Calculating the total billing amount over the entire customer tenure

dataset_enc3['Total_Billing_Amount'] = dataset_enc3['Monthly_Bill'] * dataset_enc3['Customer_tenure_months']
```

```python
# Displaying the first 5 rows of modified dataset with engineered features

print(dataset_enc3.head())
```

```python
#                                      *****STEP 3: MODEL BUILDING*****
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
# Initializing and training a Random Forest Classifier

rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)
rf_classifier.fit(X_train, y_train)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
# Initializing and training a Logistic Regression model

logistic_reg = LogisticRegression(random_state=42)
logistic_reg.fit(X_train, y_train)
```

```python
from sklearn.svm import SVC
```

```python
# Initializing and training a Support Vector Machine(SVM) classifier

svm_classifier = SVC(kernel='poly', random_state=42) #Using a polynomial kernel
svm_classifier.fit(X_train, y_train)
```

```python
#                                       *****STEP 4: MODEL OPTIMIZATION*****
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
# Defining the hyperparameter grid for RandomizedSearchCV

param_dist = {
    'C': [0.1, 1, 10],  # Regularization parameter
    'kernel': [ 'poly'],  # Kernel type
    'gamma': ['scale', 'auto', 0.1, 1],  # Kernel coefficient 
    'degree': [2, 3, 4]  # Degree of the polynomial kernel
}
```

```python
param_dist
```

```python
# Initializing RandomizedSearchCV

random_search = RandomizedSearchCV(svm_classifier, param_distributions=param_dist, n_iter=10, cv=5, n_jobs=-1, random_state=42)

# Fitting the model with optimized hyperparameters

random_search.fit(X_train, y_train)

# Getting the best hyperparameters and the best model

best_params = random_search.best_params_
best_model = random_search.best_estimator_
```

```python
# Evaluating the model

from sklearn.metrics import accuracy_score, classification_report
```

```python
# Making predictions on the test set

y_pred1 = rf_classifier.predict(X_test)
y_pred2 = logistic_reg.predict(X_test)
y_pred3 = svm_classifier.predict(X_test)
```

```python
# Calculating accuracy and other classification metrics for Random Forest Model

accuracy1 = accuracy_score(y_test, y_pred1)
classification_report_output = classification_report(y_test, y_pred1)

print(f'Accuracy: {accuracy1}')
print('Classification Report:\n', classification_report_output)
```

```python
# Calculating accuracy and other classification metrics for Logistic Regression Model

accuracy2 = accuracy_score(y_test, y_pred2)
classification_report_output = classification_report(y_test, y_pred2)

print(f'Accuracy: {accuracy2}')
print('Classification Report:\n', classification_report_output)
```

```python
# Calculating accuracy and other classification metrics for SVM Model

accuracy3 = accuracy_score(y_test, y_pred3)
classification_report_output = classification_report(y_test, y_pred3)

print(f'Accuracy: {accuracy3}')
print('Classification Report:\n', classification_report_output)
```

```python
# Making predictions with the best model

y_pred = best_model.predict(X_test)

# Evaluating the best model

accuracy = accuracy_score(y_test, y_pred)
classification_report_output = classification_report(y_test, y_pred)

print("Best Hyperparameters:", best_params)
print(f'Accuracy with Best Model: {accuracy}')
print('Classification Report with Best Model:\n', classification_report_output)
```

```python
import joblib
from flask import Flask, request, jsonify


joblib.dump(best_model, 'churn_prediction_model.pkl')

# Initializing Flask application
app = Flask(__name__)

# Loading the trained model
model = joblib.load('churn_prediction_model.pkl')  

# Defining an API route for making predictions
@app.route('/predict', methods=['POST'])
def predict():
    try:
        # Getting data from the request
        data = request.get_json()
        
        # Converting data to a DataFrame
        data_df = pd.DataFrame(data)
        
        # Making predictions
        predictions = model.predict(data_df)
        
        # Converting predictions to a list
        predicted_labels = predictions.tolist()
        
        # Returning the predicted labels as JSON response
        return jsonify({'predictions': predicted_labels})

    except Exception as e:
        return jsonify({'error': str(e)})

if __name__ == '__main__':
    # Running the Flask application on a specific host and port
    app.run(host='0.0.0.0', port=5000)
```
